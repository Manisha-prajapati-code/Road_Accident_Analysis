```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the dataset

```python
file_path = 'Road_Accident_Data.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)
```

# Data overview


```python
print("Dataset Overview:")
print(df.head())
print("\nBasic Information:")
print(df.info())
```

# Convert Date and Time to datetime for analysis

```python
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df['Hour'] = df['DateTime'].dt.hour
```

# Set up the figure and subplots

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Road Accident Data Analysis', fontsize=18)

# Plot 1: Accident distribution by time of day
sns.countplot(x='Hour', data=df, palette='viridis', ax=axes[0, 0])
axes[0, 0].set_title('Accidents by Hour of Day', fontsize=14)
axes[0, 0].set_xlabel('Hour of Day')
axes[0, 0].set_ylabel('Number of Accidents')

# Plot 2: Accident distribution by weather condition
sns.countplot(y='Weather_Condition', data=df, palette='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title('Accidents by Weather Condition', fontsize=14)
axes[0, 1].set_xlabel('Number of Accidents')
axes[0, 1].set_ylabel('Weather Condition')

# Plot 3: Accident distribution by road condition
sns.countplot(y='Road_Condition', data=df, palette='cubehelix', ax=axes[1, 0])
axes[1, 0].set_title('Accidents by Road Condition', fontsize=14)
axes[1, 0].set_xlabel('Number of Accidents')
axes[1, 0].set_ylabel('Road Condition')

# Plot 4: Accident hotspots by location
sns.countplot(y='Location', data=df, order=df['Location'].value_counts().index, palette='mako', ax=axes[1, 1])
axes[1, 1].set_title('Accident Hotspots by Location', fontsize=14)
axes[1, 1].set_xlabel('Number of Accidents')
axes[1, 1].set_ylabel('Location')

# Plot 5: Correlation between fatalities, injuries, and driver age
sns.scatterplot(x='Injuries', y='Fatalities', size='Driver_Age', hue='Driver_Age', data=df, ax=axes[2, 0], palette='cool')
axes[2, 0].set_title('Fatalities vs Injuries and Driver Age', fontsize=14)
axes[2, 0].set_xlabel('Injuries')
axes[2, 0].set_ylabel('Fatalities')

# Plot 6: Fatalities vs weather condition
sns.boxplot(x='Weather_Condition', y='Fatalities', data=df, palette='Set3', ax=axes[2, 1])
axes[2, 1].set_title('Fatalities by Weather Condition', fontsize=14)
axes[2, 1].set_xlabel('Weather Condition')
axes[2, 1].set_ylabel('Fatalities')
axes[2, 1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Add space for the title
plt.show()
```

# Road Accident Analysis

## Objective
The objective of this project is to analyze traffic accident data to:
- Identify patterns related to road conditions, weather, and time of day.
- Visualize accident hotspots and contributing factors.
- Provide actionable insights to improve road safety and reduce accidents.

---

## Problem Statement
Road accidents are a major cause of fatalities and injuries worldwide. Understanding the patterns behind accidents can help:
1. Identify high-risk times, locations, and conditions.
2. Highlight the impact of weather and road conditions.
3. Develop effective traffic management and safety strategies.

Key Challenges:
- Large volumes of accident data are often underutilized.
- Identifying critical accident hotspots and causes requires effective visualization and analysis.
- Weather and time-based correlations are not always straightforward.

---

## Solution
This project addresses the problem by:
1. **Data Preparation**:
   - Loading and cleaning the accident dataset.
   - Extracting critical features such as accident time, weather, and road conditions.

2. **Data Analysis**:
   - Identifying accident distribution across different times of the day.
   - Analyzing the impact of weather and road conditions on accidents.
   - Highlighting accident hotspots using visualizations.

3. **Visualizations**:
   - Multiple subplots for analyzing time-based, weather-based, and location-based accident patterns.
   - Use of libraries like `matplotlib` and `seaborn` for clear and insightful plots.

4. **Insights**:
   - Peak accident hours (e.g., morning and evening rush hours).
   - Increased accidents during adverse weather conditions (foggy and rainy).
   - Identification of high-risk locations for targeted interventions.
